```python
import time
import requests
from bs4 import BeautifulSoup

# Base URL of the website
base_url = "https://www.nycgovparks.org/art-and-antiquities/permanent-art-and-monuments/results"

data_all = []

# Loop through pages 1 to 3
for page_number in range(1, 4):
    print(f"Scraping page {page_number}...")
    
    # Fetch the page
    response = requests.get(base_url, params={"page": page_number})
    if response.status_code != 200:
        print(f"Failed to fetch page {page_number}")
        continue
    
    # Parse the page content
    soup_doc = BeautifulSoup(response.content, "html.parser")
    rows = soup_doc.find_all(class_='mon_list')  # Find all rows with class 'mon_list'
    
    # Process each row
    for row in rows:
        print("-----")
        print(row.text.strip())  # Debug: Show the raw text of the row
        
        data = {}  # Dictionary to store the scraped data
        
        # Extract title
        title_element = row.find('strong')
        data['title'] = title_element.text.strip() if title_element else "No title"
        
        # Extract location
        location_element = row.find('em')
        data['loc'] = location_element.text.strip() if location_element else "No location"
        
        # Extract description (last text after a <br> tag)
        desc_elements = row.find_all('br')
        if desc_elements:
            last_text = desc_elements[-1].next
            data['desc'] = last_text.strip() if last_text else "No description"
        else:
            data['desc'] = "No description"
        
        # Add the data to the list
        data_all.append(data)
        time.sleep(1)  # Pause for 1 second to be polite
    
    # Debug: Print the collected data
    for data in data_all:
        print(data)
```

```python
import time
import requests
from bs4 import BeautifulSoup

url = "https://www.nycgovparks.org/art-and-antiquities/permanent-art-and-monuments/results"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}
data_all = []
for page_number in range(1, 82):
    print(f"Scraping page {page_number}...")
    try:
        response = requests.get(url, headers=headers, params={"search_text": " ", "borough": "", "page": page_number})
        if response.status_code != 200:
            print(f"Failed to scrape page {page_number}.")
            continue
        soup_doc = BeautifulSoup(response.content, "html.parser")
        rows = soup_doc.find_all(class_='mon_list')
        for row in rows:
            data = {}
            title_element = row.find('strong')
            data['title'] = title_element.text.strip() if title_element else "No title"
            location_element = row.find('em')
            data['loc'] = location_element.text.strip() if location_element else "No location"
            desc_elements = row.find_all('br')
            if desc_elements:
                last_text = desc_elements[-1].next
                data['desc'] = last_text.strip() if last_text else "No description"
            else:
                data['desc'] = "No description"
            data_all.append(data)
        time.sleep(2)
    except Exception as e:
        print(f"An error occurred")
        continue
for data in data_all:
    print(data)
```

```python
import pandas as pd

df = pd.json_normalize(data_all)
df.head()
```
